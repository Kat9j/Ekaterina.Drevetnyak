## [Исследовательский анализ данных](#intro4)

**Посчитаем, сколько книг вышло после 1 января 2000 года**

```python
query = '''
SELECT COUNT(book_id)
FROM books
WHERE publication_date > '01.01.2000'
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

Посе 1 января 2000 года вышло 819 книг.

**Для каждой книги посчитаем количество обзоров и среднюю оценку**

```python
query = '''
SELECT b.title
      ,reviews_query.count_reviews AS count_reviews
      ,AVG(rat.rating) AS avg_rating
      
FROM books AS b

LEFT JOIN (SELECT book_id
                  ,COUNT(review_id) AS count_reviews
           FROM reviews
           GROUP BY 1
          ) AS reviews_query ON reviews_query.book_id = b.book_id
          
LEFT JOIN ratings AS rat ON rat.book_id = b.book_id

WHERE count_reviews IS NOT NULL
GROUP BY 1, 2
ORDER BY 2 DESC
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

Из 1000 книг 994 книги имеют обзоры. Самое большое колиество 7 обзоров написано на книгу Twilight (Twilight #1), средний рейтинг у неё 3,66. 6 книг не имеют обзоров.

**Определим издательство, которое выпустило наибольшее число книг толще 50 страниц — так мы исключаем из анализа брошюры**

```python
query = '''
SELECT p.publisher
       ,COUNT(b.book_id) AS count_books
FROM books AS b
LEFT JOIN publishers AS p ON p.publisher_id = b.publisher_id
WHERE num_pages > 50
GROUP BY 1
ORDER BY 2 DESC
LIMIT 1
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

Издательство, которое выпустило наибольшее число книг толще 50 страниц - это Penguin Books. Оно выпустило 42 книги.

**Определим автора с самой высокой средней оценкой книг — учитываем только книги с 50 и более оценками**

```python
query = '''
SELECT a.author
       ,AVG(ratings_query.avg_rating) AS avg_rating
FROM books AS b
LEFT JOIN (
           SELECT book_id
                  ,AVG(rating) AS avg_rating
                  ,COUNT(rating_id) AS count_rating
            FROM ratings
            GROUP BY 1
          ) AS ratings_query ON ratings_query.book_id = b.book_id
LEFT JOIN authors AS a ON a.author_id = b.author_id
WHERE count_rating >= 50
GROUP BY 1
ORDER BY 2 DESC
LIMIT 1
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

Самая высокая средняя оценка книг у автора J.K. Rowling/Mary GrandPré - Джоан К. Роулинг/Мэри Гранпре. Средняя оценка ее книг составляет 4.28 балла.

**Посчитаем среднее количество обзоров от пользователей, которые поставили больше 48 оценок.**

```python
query = '''
WITH
ratings_query AS (
                     SELECT r.username
                            ,COUNT(rating_id) AS count_ratings
                     FROM ratings AS r
                     GROUP BY 1
                     HAVING COUNT(rating_id) > 48
                     ),
reviews_query AS (
                     SELECT rev.username
                            ,COUNT(review_id) AS count_reviews
                     FROM reviews AS rev
                     GROUP BY 1
                     )
SELECT AVG(count_reviews) AS avg_reviews_count
FROM ratings_query
LEFT JOIN reviews_query ON ratings_query.username = reviews_query.username
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

Пользователи, которые поставили больше 48 оценок очень активно пишут и отзывы. В среднем ими написано 24 обзора! Настоящие книголюбы)
